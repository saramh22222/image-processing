```python
import cv2 , matplotlib.pyplot as plt , numpy as np 
from skimage.filters import threshold_otsu
```

```python
sample_img = cv2.imread("C:/Users/photo1697438595.jpeg")
img=cv2.cvtColor(sample_img ,cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(img ,cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
#applay treasholding
treash = threshold_otsu(gray)
img_otsu = gray < treash
plt.imshow(img_otsu)
```

```python
# Define a function to filter an image with a mask
def filter_image(image, mask):
    r = image[:,:,0] * mask
    g = image[:,:,1] * mask
    b = image[:,:,2] * mask
    return np.dstack([r,g,b])

filtered= filter_image(img,img_otsu)
plt.imshow(filtered)    
```
