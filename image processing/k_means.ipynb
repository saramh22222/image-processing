```python
import cv2 , numpy as np , matplotlib.pyplot as plt
```

```python
sample_image=cv2.imread("C:/Users/photo1697438595.jpeg")
```

```python
plt.imshow(sample_image)
```

```python
img=cv2.cvtColor(sample_image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
#preprocessing image 
towdimg=img.reshape((-1,3))
towdimg=np.float32(towdimg)
```

```python
#define parameter 
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER , 100 , 0.95)
k = 20
atemps = 10
```

```python
#apply k means 
Ret, Label, Center = cv2.kmeans(towdimg, k, None, criteria, atemps, cv2.KMEANS_PP_CENTERS)
```

```python
#recover img
Center=np.uint8(Center)
res = Center[Label.flatten()]
last_img = res.reshape(img.shape)
```

```python
plt.imshow(last_img)
```
